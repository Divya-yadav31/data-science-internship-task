```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2002, 2004, 2006, 2008]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
tendulkar = [500, 700, 900, 1100, 1500, 2000, 1800, 1600, 800, 0]

plt.style.use("Solarize_Light2")

plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, tendulkar, color='blue', label="Tendulkar")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

product_A = [500, 650, 800, 1000, 1300, 1600, 2000, 2500, 3000, 3500]
product_B = [400, 550, 700, 900, 1100, 1300, 1500, 1800, 2100, 2500]
product_c = [50, 100, 200, 300, 500, 700, 900, 1200, 1500, 1800]

# Use a clean style
plt.style.use("seaborn-v0_8-talk")

# Plotting each company with different styles
plt.plot(years, product_A, color='orange', linestyle='--', marker='o', label="product_A")
plt.plot(years, product_B, color='blue', linestyle='-.', marker='s', label="product_B")
plt.plot(years, product_c, color='green', linestyle='-', marker='^', label="product_C")

# Adding labels and title
plt.xlabel("Year")
plt.ylabel("Annual Sales (Millions)")
plt.title("Sales Growth (2013–2022)")

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the graph
plt.show()
```
