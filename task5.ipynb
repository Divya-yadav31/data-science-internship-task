```python
import matplotlib.pyplot  as plt
import numpy as np
```

```python
years = [2000, 2002, 2004, 2006, 2008]
kohli = [5000, 8000, 1100, 1300, 1500]
sehwag = [3000, 800, 1200, 1500, 1700]
Sachin = [1100, 15000, 2000, 1800, 1600]

x = np.arange(len(years))
width = 0.25 # it will set width

#it will define label and width of a bar graph for each player
plt.bar(x - width, Sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")

plt.xlabel("Year") # it will give nameor label to x - axis
plt.ylabel("Runs")# and this to y axis
plt.title("Run Comparison") # it will give the title to graph
plt.xticks(x, years)  # it will show real years rather than 0,1,2,..
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Kohli", "Sehwag", "Sachin"]
runs_5yrs = [5000+8000+1100+1300+1500, 3000+8000+1200+1500+1700, 1100+1500+2000+1800+1600]
plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout() 
```

 Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.


```python
import matplotlib.pyplot as plt
```

```python

os_names = ["Android" , "iOS" , "others" ]
market_share = [70 , 25 , 5]

plt.style.use("classic")
plt.title("Market Share of Mobile OS")
plt.pie(market_share, labels= os_names )
plt.show()

```

```python
os_names = ["Android" , "iOS" , "others" ]
market_share = [70 , 25 , 5]
plt.barh(os_names , market_share  , color = "blue")
plt.xlabel("Market Share of OS")
plt.title("Mobile OS ")
plt.tight_layout()
```
